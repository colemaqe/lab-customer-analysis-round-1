```python
def clean(x):
    x = x.upper()
    x = x.strip()
    if x in ['AZ', 'ARIZONA']:
        return 'AZ'
    elif x in ['CA', 'CALIFORNIA', 'CALI']:
        return 'CA'
    elif x in ['TN', 'TENNESSEE']:
        return 'TN'
    else:
        return x
```

```python
def clean(x):
    if x in ['M']:
        return 'Male'
    elif x in ['Femal', 'F' , 'female']:
        return 'Female'
    else:
        return 'U'
    
data['gender'] = list(map(clean, data['gender']))
```
